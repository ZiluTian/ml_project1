```python
import pandas as pd, matplotlib.pyplot as plt, pylab as pl, numpy as np
libPath = 'project1/scripts'
import sys
if not libPath in sys.path: sys.path.append(libPath)
from implementations import *
# Use a subset 
df = pd.read_csv("train.csv")
df.head()
```

```python
# Calculate the correlation between Prediction and various Feature 
y,inputs,id=read_clean_csv()
cross_correlate_vec = []
for i in range(29):
# Cross-correlation of two 1-dimensional sequences, c_{av}[k] = sum_n a[n+k] * conj(v[n])
    cross_correlate_vec.append(np.correlate(df.iloc[:,1], df.iloc[:,2+i])[0])

coef_vec = []
for i in range(29):
# Pearson product-moment correlation coefficients
    coef_vec.append(np.corrcoef(df.iloc[:,1], df.iloc[:,2+i])[0,1])

# print(coef_vec)
```

```python
# plt.scatter(cross_correlate_vec, coef_vec)

plt.scatter(range(29), np.abs(coef_vec))
plt.grid
plt.xlabel("Feature number")
plt.ylabel("Pearson product-moment correlation coef")
# plt.title("Pearson Correlation coefficient of features against prediction")
# plt.savefig("Figures/feature_corr_coef_2")
plt.show()
plt.savefig('fig_corr_coef1',format=eps)
# Note the difference in y-axis scale 
plt.scatter(range(29), np.abs(cross_correlate_vec))
plt.grid
plt.xlabel("Feature number")
plt.ylabel("Absolute value of cross correlation")
# plt.title("Correlation coefficient of features against prediction")
plt.show()

# print(coef_vec)
# print(cross_correlate_vec)
```

```python
# get the number of features that have coef value greater than 10 
coef_vec_abs = np.abs(coef_vec)

np.sum(coef_vec_abs>0.1)

feature_list_ind = [i for i, x in enumerate(coef_vec_abs) if x>0.1]

# feature list, 0 indexed starting from DER_mass_MMC
print(feature_list_ind)

# updated feature list when using the entire training set 
# [1, 3, 4, 5, 6, 8, 10, 11, 12, 13, 22]

#original feature list when using a sample of the training set 
# [ 1,  5,  8, 10, 11, 12, 13, 17, 22, 25]
```

```python
# correlation matrix for selected features 
feature_num = len(feature_list_ind)
corr_matrix = np.array([[0.1 for i in range(feature_num)] for j in range(feature_num)])
for i in range(feature_num): 
    foo = feature_list_ind[i]
    for j in range(feature_num): 
        bar = feature_list_ind[j]
# offset by 2, since the feature_list_ind is 0 based starting from DER_mass 
        corr_matrix[i][j]=np.corrcoef(df.iloc[:,2+foo], df.iloc[:,2+bar])[0,1]
```

```python
# print(coef_matrix)
# Feature 0-10 corresponding to feature val in feature_list_ind, not in original table
plt.imshow(np.abs(corr_matrix), interpolation='nearest')
plt.title("Correlation Coef of Selective Features")
plt.colorbar()
plt.show()
```

```python
# Correlation coefficient matrix for all features (from Prediction to the last one)
corr_matrix_full = np.array([[0.1 for i in range(31)] for j in range(31)])
for i in range(31): 
    for j in range(31): 
        corr_matrix_full[i][j]=np.corrcoef(df.iloc[:,i+1], df.iloc[:,j+1])[0,1]

plt.imshow(np.abs(corr_matrix_full), interpolation='nearest', cmap='winter')
plt.title("Correlation Coefficient for All Features")
plt.colorbar()
plt.show()  

# plt.savefig("Figures/feature_corr_coef_all")
```

```python
from numpy import linalg as LA

w,v = LA.eig(np.abs(corr_matrix_full))

print("Eigenvalues",w)
print("Eigenvectors", v)
```

```python
# Covariance coefficient matrix for all features (from Prediction to the last one)
cov_matrix_full = np.array([[0.1 for i in range(31)] for j in range(31)])
for i in range(31): 
    for j in range(31): 
        cov_matrix_full[i][j]=np.cov(df.iloc[:,i+1], df.iloc[:,j+1])[0,1]

plt.imshow(np.abs(cov_matrix_full), interpolation='nearest', cmap='winter')
plt.title("Covariance Coefficient for All Features")
plt.colorbar()
plt.show()  

w,v = LA.eig(np.abs(cov_matrix_full))

print("Eigenvalues",w)
print("Eigenvectors", v)

# cross_correlate_vec = []
# for i in range(29):
#     cross_correlate_vec.append(np.correlate(df.iloc[:,1], df.iloc[:,2+i])[0])

# plt.imshow(np.abs(corr_matrix_full), interpolation='nearest', cmap='winter')
# plt.colorbar()
# plt.show()  
# plt.savefig("Figures/feature_corr_coef_all")

# print(np.cov(df.iloc[:,1], df.iloc[:,3])[0,1])
```

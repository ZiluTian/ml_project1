```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('log_file.txt') as f:
    file11= f.readlines()
```

```python
av_tr=[l[19:-2].split(',') for l in file11 if l[:17]=='Average_acc_train']
```

```python
av_te=[l[18:-2].split(',') for l in file11 if l[:16]=='Average_acc_test']
```

```python
av_te[1]=[float(av_te[1][i]) for i in np.arange(0,len(av_te[1]),2)]
av_tr[1]=[float(av_tr[1][i]) for i in np.arange(0,len(av_tr[1]),2)]
```

```python
av_te[2]=[float(av_te[2][i]) for i in np.arange(0,len(av_te[2]),1)]
av_tr[2]=[float(av_tr[2][i]) for i in np.arange(0,len(av_tr[2]),1)]
```

```python
av_te[4]=[float(av_te[4][i]) for i in np.arange(0,len(av_te[4]),2)]
av_tr[4]=[float(av_tr[4][i]) for i in np.arange(0,len(av_tr[4]),2)]
```

```python
plt.figure(figsize=(5,4))
plt.plot(np.arange(20)+1, av_tr[1], '-o', color = 'b')
plt.plot(np.arange(20)+1,av_te[1],'-o',color='r')
plt.xticks(np.arange(0, 21, 2))
plt.ylim([30, 100])
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend(['Train', 'Test'], loc='lower right')
plt.savefig('figs/8_0.eps')
```
